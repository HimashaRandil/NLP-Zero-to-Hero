```python
import wave
import matplotlib.pyplot as plt
import numpy as np
```

```python
obj = wave.open("voice.wav", "rb")
```

```python
sample_fre = obj.getframerate()
n_sample = obj.getnframes()
signal_wave = obj.readframes(-1)
```

```python
obj.close()
```

```python
t_audio = n_sample/sample_fre
```

```python
print(t_audio, "seconds")
```

```python
signal_array = np.frombuffer(signal_wave, dtype=np.int16)
print(signal_array.shape)
```

```python
signal_length = len(signal_array)
```

```python
times = np.linspace(0, (n_sample - 1) / sample_fre, num=signal_length)
```

```python
print(times.shape)
```

```python
plt.figure(figsize=(15, 5))
plt.plot(times, signal_array)
plt.title('Audio')
plt.ylabel('Signal Value')
plt.xlabel('Time (s)')
plt.xlim(0, t_audio)
plt.show()
```
